```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Round-Robin analysis

```python
cpu_bound = pd.read_csv("Round-Robin/CPU_bound.csv")

print("CPU-Bound - Round Robin")
cpu_bound
```

```python
scpu_bound = pd.read_csv("Round-Robin/SCPU_bound.csv")

print("SCPU-Bound - Round Robin")
scpu_bound
```

```python
io_bound = pd.read_csv("Round-Robin/IO_bound.csv")

print("I/O-Bound - Round Robin")
io_bound
```

```python
quantum   = cpu_bound["INTERV"]
cpu_turn  = cpu_bound["AvgTurn"]
scpu_turn = scpu_bound["AvgTurn"]
io_turn   = io_bound["AvgTurn"]

plt.title("Round Robin for different quantum")
plt.ylabel("Turnaround average")
plt.xlabel("Quantum")
plt.xticks(ticks=quantum)

plt.plot(quantum, cpu_turn, label="CPU-Bound")
plt.plot(quantum, scpu_turn, label="SCPU-Bound")
plt.plot(quantum, io_turn, label="I/O-Bound")
plt.legend()
plt.show()
```
